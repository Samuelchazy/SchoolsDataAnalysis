# illustrate Subjects(math,science,...) Distribution per school

```python
def subject_distribution(X,Y,w,h,n_plots_y,n_plots_x,color):
    # X is the feature(math,science,...) & Y is the school name
    school = df_compact['SchName'].where(df_compact['SchName']==Y)
    fig,ax = plt.subplots(n_plots_x,n_plots_y,figsize=(w,h),dpi=300,sharex=False,sharey=True)
    
    i=0
    for n_x in range(n_plots_x):
        for n_y in range(n_plots_y):
            if i < len(X):
                feature = df_compact[X[i]].where(df_compact['SchName']==Y).dropna()
                plot = sns.histplot(ax=ax[n_x][n_y],x=feature,palette='flare',color=color,alpha=0.3)
                i+=1
            else:
                break;
```

# find the name of the school

```python
def find_school(name):
    #name = name.lower()
    school = df_compact['SchName']
    school_name = []
    for sc in school:
        if name in sc:
            school_name.append(sc)
    
    if len(school_name) > 0:
        return set(school_name)
    else:
        return('No match found for this school!')
    
```

# Compare the selcted subject to the same subject in all other schools

```python
def compare_subject_inschool_vs_other_schools(X,Y):
    # X is subject(math, science,...) & Y is school name
    subject = df_compact[X].where(df_compact['SchName'] == Y).dropna()
    print(f'The average for({Y}) in ({X}) is: {round(subject.mean(),2)}')
    subject_all = df_compact[X]
    print(f'The average for all schools in ({X}) is: {round(subject_all.mean(),2)}')
        
    fig,ax = plt.subplots(figsize=(15,5),dpi=300)
    ax_one = sns.barplot(y=subject,color='blue',alpha=0.3)
    ax_two = sns.barplot(y=subject_all,color='magenta',alpha=0.3)
    
    names = ['Your school','All schools']
    i = 0
    for x in ax.patches:
        ax.annotate(round(x.get_height(),2),(x.get_x()-0.01,x.get_height()),ha='right',
                    va='top',fontsize=10,color='grey')
        plt.text(-0.39,x.get_height(),names[i],ha='left',fontsize=10,color='black')
        i += 1
    
    return
```

# Get the confidence interval for the selected subject

```python
from scipy import stats
import statsmodels.api as sm

subject = 'SciPV2'

n = len(df_compact[subject])
X_bar = df_compact[subject].mean()
X_bar_median = df_compact[subject].median()
std = df_compact[subject].std()
std_error = std / np.sqrt(n)
Z_multiplier = stats.norm.ppf(0.95+(0.05/2))

# CI_low = X_bar - Z_multiplier * std_error
# CI_high = X_bar + Z_multiplier * std_error
CI_interval = stats.norm.interval(alpha=0.95,loc=X_bar,scale=std_error)

print(f'The average mean for this subject is {X_bar}')
print(f'The average median for this subject is {X_bar_median}')
print(f'The standard error for this subject is {std_error}')
print(f'The 95% confidence interval for this subject average is {CI_interval}')
```

# Build a machine learning model to predict the levels

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier,RandomForestRegressor
from sklearn.metrics import confusion_matrix,classification_report,plot_roc_curve,plot_precision_recall_curve
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.tree import plot_tree
```

# Let us find the best tree depth

```python
def depth_score(md):
    sub = df_school_subjects
    
    # Split the data
    X_train,X_test,y_train,y_test = train_test_split(sub,target_two,
                                        test_size=0.25,stratify=target_two)
    
    # Scale down the data
    scaler = StandardScaler()
    X_train = scaler.fit_transform(X_train)
    X_test = scaler.transform(X_test)
    
    # Apply dimensionality reduction to the data using PCA
    pca = PCA()
    X_train = pca.fit_transform(X_train,y_train)
    X_test = pca.transform(X_test)
    
    model = RandomForestClassifier(max_depth=md)
    model.fit(X_train,y_train)
    train_score = model.score(X_train,y_train)
    test_score = model.score(X_test,y_test)
    return {'Max Depth: ':md,'training score: ':train_score,'testing score: ':test_score}

score_df = pd.DataFrame([depth_score(md) for md in range(1,15)])
score_df.head(20)
```

# Let us find the best tree leaf node

```python
def leaves_score(md):
    sub = df_school_subjects
    
    # Split the data
    X_train,X_test,y_train,y_test = train_test_split(sub,target_two,
                                        test_size=0.25,stratify=target_two)
    
    # Scale down the data
    scaler = StandardScaler()
    X_train = scaler.fit_transform(X_train)
    X_test = scaler.transform(X_test)
    
    model = RandomForestClassifier(max_leaf_nodes=md)
    model.fit(X_train,y_train)
    train_score = model.score(X_train,y_train)
    test_score = model.score(X_test,y_test)
    return {'Max Depth: ':md,'training score: ':train_score,'testing score: ':test_score}

score_df = pd.DataFrame([leaves_score(md) for md in range(2,30)])
score_df.head(20)
```

```python
def predict_level():
    sub = df_school_subjects
    
    # Split the data
    X_train,X_test,y_train,y_test = train_test_split(sub,target_two,
                                        test_size=0.25,stratify=target_two)
    
    # Scale down the data
    scaler = StandardScaler()
    X_train = scaler.fit_transform(X_train)
    X_test = scaler.transform(X_test)
    
    # Build the Decision Tree Classifer Model
    model = RandomForestClassifier(n_estimators=100,max_depth=11,max_leaf_nodes=20,
                                   max_features=10,random_state=2)
    model.fit(X_train,y_train)
    y_pred = model.predict(X_test)
    
    #print('-'*85)
    importance_df = pd.DataFrame({'feature':sub.columns,
                                  'importance':model.feature_importances_}).sort_values('importance',
                                                                                        ascending=False)
    ax,fig=plt.subplots(figsize=(15,2),dpi=300)
    sns.barplot(data=importance_df.head(5),x='importance',y='feature',palette='Pastel1')
    #print('-'*85)

    print('-'*85)
    print(f'Model score on the training data: {model.score(X_train,y_train)}')
    print(f'Model score on the testing data: {model.score(X_test,y_test)}')
    
    print('-'*85)
    print(classification_report(y_test,y_pred,zero_division=0))
    
    print('-'*85)
    print('Confusion Matrix')
    print('-'*85)
    cm = confusion_matrix(y_test,y_pred)
    print(cm)
    print('-'*85)
    
    print('-'*85)
    tp = cm[0,0]
    fn = cm[0,1]
    fp = cm[1,0]
    tn = cm[1,1]
    print(f'tp: {tp} / fn: {fn} / fp: {fp} / tn: {tn}')
    
    print('-'*85)
    TPR = tp / (tp+fn)
    print(f'The True Positive Rate is: {TPR}')
    FPR = fp / (fp+tn)
    print(f'The False Positive Rate is: {FPR}')
    
    print('-'*85)
    plot_roc_curve(model,X_test,y_test)
    plt.title('ROC Curve')
    plot_precision_recall_curve(model,X_test,y_test)
    plt.title('Precision VS Recall')

    ax,fig = plt.subplots(figsize=(15,1),dpi=300)
    ax = sns.countplot(y=y_pred,palette='Pastel1',saturation=0.6)
    
```
